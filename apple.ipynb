```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('AppleStore (2).csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
free=df[df['price']==0]
free.head()
```

```python
g=sns.pairplot(free)
plt.savefig("free_pairplot.png",bbox_inches="tight")
```

```python
not_free=df[df['price']!=0]
not_free.head()
```

```python
sns.pairplot(not_free)
plt.savefig("not_free_pairplot.png",bbox_inches="tight")
```

```python
top15=df.sort_values(by=['rating_count_tot']).tail(15)
top15.head()
```

```python
df.nunique()
```

```python
top20=df.sort_values(by=['user_rating']).tail(20)
top20.head()
```

```python
fig, axes = plt.subplots(1,2,figsize=(16,6))

axes[0].barh(top20['track_name'],top20['user_rating'])
axes[0].set_title("top 20 user rating")
axes[0].set_xlabel("rating")
axes[0].set_ylabel("names")


axes[1].barh(top15['track_name'],top15['rating_count_tot'],color='red')
axes[1].set_title("top 15 rating count")
axes[1].set_xlabel("count")
axes[1].set_ylabel("names")
fig.tight_layout()
plt.savefig("top 15&20.png",bbox_inches="tight")
```

```python
fig,axes=plt.subplots(figsize=(6,6))

axes.hist(df['price'],200)
axes.set_xlim(0,30)
axes.set_title("distribution of apps prices",size=20)
axes.set_xlabel("price")
axes.set_ylabel("count")
plt.savefig("distribution of apps prices.png",bbox_inches="tight")
```

```python
sns.distplot(df['user_rating'],kde=False,color='green',label='user rating')
sns.distplot(df['user_rating_ver'],kde=False,color='red',label='user rating ver')
plt.ylabel('count')
plt.title('disribution of user rating',size=20)
plt.legend()
plt.savefig("disribution of user rating.png",bbox_inches="tight")
#distrebution of 'user rating' in the last verision vs distrebution of 'user rating' of all  the verisions
```

```python
#axes=sns.violinplot(x='user_rating',y='lang.num',data=df)
axes=sns.boxplot(x="user_rating", y="lang.num", data=df)
axes.set_ylim(0,28)
plt.savefig("langnum_user_rating.png",bbox_inches="tight")
```

```python
df['prime_genre'].unique()
Games=df[df['prime_genre']=='Games']
Shopping=df[df['prime_genre']=='Shopping']
Sports=df[df['prime_genre']=='Sports']
Photo=df[df['prime_genre']=='Photo & Video']
```

```python

other=df['prime_genre'].count()-Games['prime_genre'].count()-Shopping['prime_genre'].count()-Sports['prime_genre'].count()-Photo['prime_genre'].count()
plt.figure(figsize=(8,8))
plt.pie([Games['prime_genre'].count(),Shopping['prime_genre'].count(),Sports['prime_genre'].count(),Photo['prime_genre'].count(),other],
        labels=['Games','Shopping','Sports','Photo & Video','other'], colors=['#b8f567','#fb9d8e','#78F6FF','#3B73FF','#FFD06B'],autopct='%1.1f%%')
plt.title('the ratio between apps genres',size=20,color='purple')
plt.savefig("the ratio between apps genres.png",bbox_inches="tight")
```

```python
plt.figure(figsize=(8,8))
plt.pie([free['price'].count(),not_free['price'].count()],
        labels=['free','not free'], colors=['#b8f567','#fb9d8e'],autopct='%1.1f%%')
plt.title('the ratio between "free" apps and "not free" apps',size=20,color='purple')
plt.savefig("the ratio between free apps and not free apps.png",bbox_inches="tight")
```

```python
prime_genre_rating_means = df[['prime_genre', 'user_rating']].groupby('prime_genre').mean()
prime_genre_rating_means_sorted=prime_genre_rating_means['user_rating'].sort_values(ascending=False)
genre=prime_genre_rating_means_sorted.index
means=prime_genre_rating_means_sorted.values
xpos=np.arange(len(genre))
plt.figure(figsize=(8,8))
plt.bar(xpos,means)
plt.xticks(xpos,genre,rotation=90)
plt.title('the mean user rating of each gnre',size=20,color='purple')
plt.ylabel('user rating',size=20)
plt.savefig("the mean user rating of each gnre.png",bbox_inches="tight")
```

```python
prime_genre_price_means = df[['prime_genre', 'price']].groupby('prime_genre').mean()
prime_genre_price_means_sorted=prime_genre_price_means['price'].sort_values(ascending=False)
prime_genre_price_means_sorted
```

```python
ax=sns.barplot(x=prime_genre_price_means_sorted.index,y=prime_genre_price_means_sorted.values)
ax.set_xticklabels(prime_genre_price_means_sorted.index,rotation=90)
ax.set_ylabel('price-$',size=15)
ax.set_title('the mean price of each genre',size=20,color='blue')
plt.savefig("the mean price of each genre.png",bbox_inches="tight")
```
